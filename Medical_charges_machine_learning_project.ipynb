```python
import pandas as pd
```

```python
medical_charges_url = 'https://raw.githubusercontent.com/JovianML/opendatasets/master/data/medical-charges.csv'
```

Download the data set using the *urlretrieve* function from *urllib.request*

```python
from urllib.request import urlretrieve
```

```python
urlretrieve(medical_charges_url, 'medicalcharges.csv')
```

We can create a pandas dataframe using the downloaded file, to view and analyze

```python
medical_df = pd.read_csv('medicalcharges.csv')
```
